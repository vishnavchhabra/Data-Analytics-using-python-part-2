```python
import pandas as pd
```

```python
sales=pd.read_csv(r"C:\Users\91858\Downloads\Data Fiesta 2.0 Round 2 Docket\Data Fiesta 2.0 Round 2 Docket\Data Fiesta 2.0 Round 2 Docket\Data Fiesta 2.0 Round 2 -  Data\SALESDETAILS.csv")
```

```python
rep=pd.read_csv(r"C:\Users\91858\Downloads\Data Fiesta 2.0 Round 2 Docket\Data Fiesta 2.0 Round 2 Docket\Data Fiesta 2.0 Round 2 Docket\Data Fiesta 2.0 Round 2 -  Data\SALESREP.csv")
```

```python
df = pd.merge(sales, rep ,on="Sales Rep")
df
```

```python
df['Sales Price']=df['Sales Price'].fillna(0)
df['Sales Quantity']=df['Sales Quantity'].fillna(0)
```

```python
TotalSale= df['Sales Price']*df['Sales Quantity']
```

```python
df['Total Sale']=TotalSale
```

```python
grouped_df = df.groupby(['Sales Rep Name'])
```

```python
grouped_and_summed = grouped_df.sum()
```

```python
grouped_and_summed = grouped_and_summed.reset_index()
grouped_and_summed['Total Sale']
```

```python
df_final=grouped_and_summed.filter(['Sales Rep Name','Total Sale'], axis=1)
```

```python
df_final
```

```python
df_final.sort_values('Total Sale',ascending=False, inplace=True)
```

```python
df_final
```

```python
df_final.to_csv('salebyrep.csv')
```
